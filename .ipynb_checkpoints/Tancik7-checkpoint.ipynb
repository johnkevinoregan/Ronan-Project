```julia
function center_square_crop(img, side)
    h, w = size(img)        # height, width

    # Ensure requested side does not exceed image bounds
    side = min(side, h, w)

    # Compute top-left corner of the crop
    y0 = Int((h - side) ÷ 2) + 1
    x0 = Int((w - side) ÷ 2) + 1

    return img[y0 : y0+side-1, x0 : x0+side-1]
end
```

```julia
# ╔═╡ de5d4fa2-cd39-43a1-96ca-d5773e6ff137
begin
	struct FeatureMap
	    B::Matrix{Float32} # Gaussian matrix
		filter::Function   # fourier or identity fn
	end
	
	# Constructor
	function FeatureMap(input_dim::Int, mapping_size::Int, 
							σ::Float32, filter::Function)
	    # mapping_size corresponds to the output dimension.
	    # Since we concat sin and cos, we need B to have half that many rows.
	    b_rows = div(mapping_size, 2) #KOR: NB if mapping_size is odd, we get an error 
	    
	    # Initialise B from Gaussian distribution * SD
	    B = randn(Float32, b_rows, input_dim) .* σ
	    return FeatureMap(B, filter)
	end
	
	filter_ident(x) = Float32.(vcat(identity.(x), identity.(x)))
	filter_ft(x) = Float32.(vcat(cos.(x), sin.(x)))

	# Forward Pass
	function (f::FeatureMap)(x)
	    projection = 2f0 * π .* (f.B * x)
	    return f.filter(projection)
	end
end
```

```julia
using Flux
# ╔═╡ b448a84e-58df-41d0-8d35-c1915618b4a7
begin
	# Hyperparameters
	mapping_size = 6 # KOR: I think this has to be even !!!
	#KOR: if I put mapping size = 2 or 6 or 4 or 100 this works. Just 3 or 5 doesnt work!!!
	σ = 1.0f0
	
	# Define the Model
	model = Chain(
	    FeatureMap(2, mapping_size, σ, filter_ident), # Inputs: 2 -> 256
	    Dense(mapping_size, 256, relu),
#	    Dense(256, 256, relu),
#	    Dense(256, 256, relu),
	    Dense(256, 3, sigmoid) # Output 3 (RGB), sigmoid forces 0-1 range
	)
end
```

```julia
# ╔═╡ 1dab9635-9a35-49d9-833b-ae5732b66fc8
function loss(m, X, y)
	return mean((m(X) .- y).^2)
end

using JLD2

# ╔═╡ 9804fb53-6eca-4231-a55c-2261566eaf89
function train_the_model(model, X, y, nepoch)
	opt = Flux.Adam(1f-4)
    opt_state = Flux.setup(opt, model)

    losses = Float32[]
    for epoch in 1:nepoch
        grads = Flux.gradient(model) do m
            loss(m, X, y)
        end
        Flux.Optimise.update!(opt_state, model, grads[1])
		loss_val = loss(model, X, y)
        push!(losses, loss_val)
		@info "Epoch $epoch; Loss: $loss_val"
		if epoch % 100 == 0
			model_state = Flux.state(model)
			@save "fox_FT_$(epoch).jld2" model_state opt_state epoch losses
			@info "File fox_FT_$(epoch).jld2 saved"
		end
    end
	return losses
end
```

```julia
# ╔═╡ bfd77849-cf62-48fc-9adf-a1d7627ac946
nepochs = 10

# ╔═╡ efb01834-10e1-4876-b529-ae56480873ae
losses = train_the_model(model, X_train, Y_train, nepochs)

# ╔═╡ 8fa4e47e-b118-4120-b92a-99b96cd125c2
# ╠═╡ disabled = true
#=╠═╡
losses = train_the_model(model, X_train_sub, Y_train_sub, nepochs)
  ╠═╡ =#

# ╔═╡ 5dcb4fc2-3223-41cd-a4a0-fd55badf73d3
let 
    f = Figure()
    ax = CairoMakie.Axis(f[1,1], xlabel="Epoch", ylabel="MSE")
    lines!(ax, losses)
    f
end
```
